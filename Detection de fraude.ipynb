```python
# Resample non-fraudulent transactions without replacement to match the number of instances in fraudulent class
# plus de détail : It samples (selects) a number of non-fraudulent transactions equal to the number of fraudulent ones (fraud_count) from the non_fraudulent_transactions dataset.
resampled_non_fraudulent_transactions = non_fraudulent_transactions.sample(n=fraud_count, random_state=34)
```

# Random Forest Classifier


```python
def train_random_forest(X_train, y_train, X_dev, y_dev):
    rf = RandomForestClassifier(random_state=34)
    param_grid = {
        'n_estimators': [50, 100, 200], #nombre d'arbre
        'max_depth': [None, 10, 20],  #profondeur
        'min_samples_split': [2, 5, 10]  
    }
    grid_search = GridSearchCV(rf, param_grid, scoring='recall', n_jobs=-1)
    grid_search.fit(X_dev, y_dev)
    best_rf = grid_search.best_estimator_
    best_rf.fit(X_train, y_train)
    return best_rf
```

```python
from sklearn.metrics import classification_report, roc_auc_score, precision_score, recall_score, f1_score
from tabulate import tabulate

from sklearn.metrics import classification_report, roc_auc_score, precision_score, recall_score, f1_score
from tabulate import tabulate
import warnings
import numpy as np

# Generate classification report
def generate_report(models, X_test, y_test):
    report_data = []
    for name, model in models.items():
        y_pred = model.predict(X_test)
        try:
            # Use try-except to catch warnings
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")  # Suppress warnings
                report = classification_report(y_test, y_pred, output_dict=True)
        except Warning as w:
            print(f"Warning encountered for model {name}: {str(w)}")
            report = {}  # Empty report
        auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
        report_data.append({
            "Model": name,
            "Precision": precision_score(y_test, y_pred),
            "Recall": recall_score(y_test, y_pred),
            "F1-Score": f1_score(y_test, y_pred),
            "Support": np.sum(y_pred),
            "AUC": auc
        })
    return tabulate(report_data, headers="keys", tablefmt="pretty")

input_models = {'Random Forest': best_rf}

report_table = generate_report(input_models, X_test, y_test)
print("Classification Report Table:")
print(report_table)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve


def plot_roc(input_models, input_model_names, X_test, y_test):
    # Suppress warnings
    warnings.filterwarnings("ignore")
    
    plt.figure(figsize=(10, 8))
    for model, name in zip(input_models, input_model_names):
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test)[:, 1]
            fpr, tpr, thresholds = roc_curve(y_test, y_prob)
            auc = roc_auc_score(y_test, y_prob)
            plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')
    plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_roc([best_rf],
         ['Random Forest'],
         X_test, y_test)
```

# Now lets use SVC (support vector classifier) which is a type of SVM used specifically for classification as it is the case in our work today

```python
def train_svc(X_train, y_train, X_dev, y_dev):
    # Scale the input features
    scaler = StandardScaler()
    warnings.filterwarnings("ignore", message="X has feature names, but SVC was fitted without feature names")
    X_train_scaled = scaler.fit_transform(X_train)
    X_dev_scaled = scaler.transform(X_dev)
    
    # Train SVC on the training set
    svc = SVC(random_state=34, probability=True)
    svc.fit(X_train_scaled, y_train)
    
    # Define the hyperparameter grid for fine-tuning
    param_grid = {
        'C': [0.1, 1, 10],
        'gamma': ['scale', 'auto'],
        'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
    }
    
    # Perform grid search with cross-validation on the test set
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        grid_search = GridSearchCV(svc, param_grid, cv=5, scoring='recall', n_jobs=-1)
        grid_search.fit(X_dev_scaled, y_dev)
    
    best_svc = grid_search.best_estimator_
    
    return best_svc
```

```python
# Train the models using the training and development sets
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

best_svc = train_svc(X_train, y_train, X_dev, y_dev)
```

```python
svc_score = best_svc.score(X_test, y_test)
print("Support Vector Classifier Accuracy:", svc_score)
```

# In addition of random forest and SVC previously, we gonna try other classifiers : Gradient Boosting, K-Nearest Neighbors (KNN), Logistic Regression

```python
#Gradient Boosting Classifier

def train_gradient_boosting(X_train, y_train, X_dev, y_dev):
    gb = GradientBoostingClassifier(random_state=34)
    param_grid = {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.1, 0.05, 0.01],
        'max_depth': [3, 4, 5]
    }
    grid_search = GridSearchCV(gb, param_grid, scoring='recall', n_jobs=-1)
    grid_search.fit(X_dev, y_dev)
    best_gb = grid_search.best_estimator_
    best_gb.fit(X_train, y_train)
    return best_gb
```

```python
#K Neighbours Classifier

from sklearn.neighbors import KNeighborsClassifier

def train_knn(X_train, y_train, X_dev, y_dev):
    knn = KNeighborsClassifier()
    param_grid = {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    }
    grid_search = GridSearchCV(knn, param_grid, scoring='recall', n_jobs=-1)
    grid_search.fit(X_dev, y_dev)
    best_knn = grid_search.best_estimator_
    best_knn.fit(X_train, y_train)
    return best_knn
```

```python
#Logistic Regression Classifier

def train_logistic_regression(X_train, y_train, X_dev, y_dev):
    # Train logistic regression on the training set
    lr = LogisticRegression(random_state=34, solver='lbfgs', max_iter=1000)
    lr.fit(X_train, y_train)
    
    # Define the hyperparameter grid for fine-tuning
    param_grid = {
        'C': [0.1, 1, 10],
        'penalty': ['l2']
    }
    
    # Perform grid search with cross-validation on the development set
    grid_search = GridSearchCV(lr, param_grid, cv=5, scoring='recall', n_jobs=-1)
    grid_search.fit(X_dev, y_dev)
    
    best_lr = grid_search.best_estimator_
    
    return best_lr
```

```python
# Train the models using the training and development sets

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

best_gb = train_gradient_boosting(X_train, y_train, X_dev, y_dev)
best_lr = train_logistic_regression(X_train, y_train, X_dev, y_dev)
best_knn = train_knn(X_train, y_train, X_dev, y_dev)
```

```python
#Evaluate models on the test set

gb_score = best_gb.score(X_test, y_test)
lr_score = best_lr.score(X_test, y_test)
knn_score = best_knn.score(X_test, y_test)
```
